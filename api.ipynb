```python
    # Import thư viện cần thiết
import requests
import pandas as pd
import numpy as np
import json
```

```python
# URL API SpaceX (các lần phóng trong quá khứ)
url = "https://api.spacexdata.com/v4/launches/past"

# Gửi yêu cầu GET đến API và nhận dữ liệu JSON
response = requests.get(url)

# Chuyển JSON sang Python dict/list
data = response.json()

# In ra một phần để xem cấu trúc
print(data[0])
```

```python
# Chuyển JSON sang DataFrame phẳng (bảng)
df = pd.json_normalize(data)

# Hiển thị 5 dòng đầu
df.head()
```

```python
# Kiểm tra thông tin cột và giá trị null
df.info()
df.isnull().sum()
```

```python
# Lọc chỉ những lần phóng có tên chứa "Falcon 9"
df = df[df['name'].str.contains("Falcon 9", case=False, na=False)]

# Reset lại chỉ mục
df = df.reset_index(drop=True)

# Kiểm tra lại kết quả
df[['name', 'rocket']].head()
```

```python
import requests
import pandas as pd

# Lấy danh sách rocket từ API
rockets_url = "https://api.spacexdata.com/v4/rockets"
rockets = requests.get(rockets_url).json()

rockets_df = pd.json_normalize(rockets)
rockets_df[["id", "name"]]
```

```python
# Lấy dữ liệu launch
launches_url = "https://api.spacexdata.com/v4/launches/past"
launches = requests.get(launches_url).json()
launches_df = pd.json_normalize(launches)

# Lọc các dòng có rocket ID trùng với Falcon 9
falcon9_id = "5e9d0d95eda69973a809d1ec"
falcon9_df = launches_df[launches_df["rocket"] == falcon9_id].reset_index(drop=True)

falcon9_df[["name", "date_utc", "rocket"]].head()
```

```python
# Kiểm tra cột nào có giá trị null
df.isnull().sum()

# Nếu có cột 'payload_mass_kg' (hoặc tương tự), thay null bằng giá trị trung bình
if 'payload_mass_kg' in df.columns:
    mean_mass = df['payload_mass_kg'].mean()
    df['payload_mass_kg'].fillna(mean_mass, inplace=True)

# Giữ nguyên các cột như 'landingpad' nếu null (vì không dùng bãi đáp)
df.head()
```

```python
# Chọn các cột quan trọng phục vụ phân tích
selected_columns = ["name", "date_utc", "rocket", "success", "details", "links.webcast"]
df_cleaned = df[selected_columns]

# Hiển thị bảng kết quả cuối
df_cleaned.head(10)
```

```python
# Lưu dữ liệu đã xử lý ra file CSV
df_cleaned.to_csv("spacex_cleaned_data.csv", index=False)

print("Dữ liệu sạch đã lưu thành công: spacex_cleaned_data.csv")
```
