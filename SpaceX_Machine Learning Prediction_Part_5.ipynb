```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Preprocessing allows us to standardize our data
from sklearn import preprocessing

# Allows us to split our data into training and testing data
from sklearn.model_selection import train_test_split

# Allows us to test parameters of classification algorithms and find the best one
from sklearn.model_selection import GridSearchCV

# Logistic Regression classification algorithm
from sklearn.linear_model import LogisticRegression

# Support Vector Machine classification algorithm
from sklearn.svm import SVC

# Decision Tree classification algorithm
from sklearn.tree import DecisionTreeClassifier

# K Nearest Neighbors classification algorithm
from sklearn.neighbors import KNeighborsClassifier
```

```python
def plot_confusion_matrix(y, y_predict):
    "This function plots the confusion matrix"
    from sklearn.metrics import confusion_matrix

    cm = confusion_matrix(y, y_predict)
    ax = plt.subplot()
    sns.heatmap(cm, annot=True, ax=ax)  # annot=True để hiển thị số trong ô
    ax.set_xlabel('Predicted labels')
    ax.set_ylabel('True labels')
    ax.set_title('Confusion Matrix')
    ax.xaxis.set_ticklabels(['did not land', 'land'])
    ax.yaxis.set_ticklabels(['did not land', 'landed'])
```

```python
import pandas as pd

URL1 = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv"
URL2 = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_3.csv"
```

```python
data = pd.read_csv(URL1)
X = pd.read_csv(URL2)
```

```python
display(data.head())
display(X.head(100))
```

```python
Task 1: Create a NumPy array from the column Class in data
Applying the method to_numpy() then assign it to the variable Y
# Nhiệm vụ 1: Tạo một mảng NumPy từ cột Class trong data
Áp dụng phương thức to_numpy() rồi gán nó cho biến Y
```

```python
Y = data['Class'].to_numpy()

print(Y[:10])  
```

```python
Task 2: Standardize the data
# Nhiệm vụ 2: Chuẩn hóa dữ liệu
```

```python

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X = scaler.fit_transform(X)
print(X.shape)
print(X[:5])
```

```python
TASK3:  create training and testing data
Use the function train_test_split to split the data X and Y into training and test data. Set the parameter test_size to 0.2 and random_state to 2.
The training data and test data should be assigned to the following labels: X_train, X_test, Y_train, Y_test
# NHIỆM VỤ 3: Tạo dữ liệu huấn luyện và kiểm tra
Sử dụng hàm train_test_split để chia dữ liệu X và Y thành dữ liệu huấn luyện và dữ liệu kiểm tra. Đặt tham số test_size thành 0,2 và random_state thành 2.
Dữ liệu huấn luyện và dữ liệu kiểm tra nên được gán các nhãn sau: X_train, X_test, Y_train, Y_test
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=2
)

print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("Y_train shape:", Y_train.shape)
print("Y_test shape:", Y_test.shape)
```

```python
Y_test.shape
```

```python
TASK 4: Find Hyperparameter
Create a Logistic Regression object.
Create a GridSearchCV object (logreg_cv) with cv = 10.
Fit the model to find the best parameters from the dictionary parameters.
# NHIỆM VỤ 4: Tìm Siêu tham số
Tạo một đối tượng Hồi quy Logistic.
Tạo một đối tượng GridSearchCV (logreg_cv) với cv = 10.
Điều chỉnh mô hình để tìm ra các tham số tốt nhất từ ​​các tham số từ điển.
```

```python
parameters = {
    'C': [0.01, 0.1, 1],
    'penalty': ['l2'],
    'solver': ['lbfgs']
}
```

```python
# Create Logistic Regression object
# Tạo đối tượng Hồi quy Logistic
lr = LogisticRegression(max_iter=1000)

# Create GridSearchCV object with 10-fold cross-validation
# Tạo đối tượng GridSearchCV với xác thực chéo 10 lần
logreg_cv = GridSearchCV(lr, parameters, cv=10)

# Fit the model to find the best parameters
# Điều chỉnh mô hình để tìm ra các thông số tốt nhất
logreg_cv.fit(X_train, Y_train)
```

```python
print("Tuned hyperparameters (best parameters):", logreg_cv.best_params_)
print("Best accuracy:", logreg_cv.best_score_)
```

```python
#TASK 5: Calculate Logistic Regression Accuracy on Test Data
#NHIỆM VỤ 5: Tính toán độ chính xác hồi quy logistic trên dữ liệu thử nghiệm
# Calculate accuracy on the test set
# Tính độ chính xác trên tập kiểm tra
accuracy_logreg = logreg_cv.score(X_test, Y_test)
print("Logistic Regression Test Accuracy:", accuracy_logreg)
```

```python
#TASK 6: Support Vector Machine (SVM) Hyperparameter Tuning
#NHIỆM VỤ 6: Điều chỉnh siêu tham số máy vectơ hỗ trợ (SVM)
from sklearn.svm import SVC

# Create SVM model
svm = SVC()

# Define hyperparameter grid
# Xác định lưới siêu tham số
parameters_svm = {
    'C': [0.1, 1, 10, 100],
    'kernel': ['linear', 'rbf', 'poly'],
    'gamma': ['scale', 'auto']
}

# Create GridSearchCV object with 10-fold CV
# Tạo đối tượng GridSearchCV với CV gấp 10 lần
svm_cv = GridSearchCV(svm, parameters_svm, cv=10)

# Fit on training data
svm_cv.fit(X_train, Y_train)

# Display best parameters and accuracy
# Hiển thị các thông số và độ chính xác tốt nhất
print("Best SVM parameters:", svm_cv.best_params_)
print("Best SVM score (CV accuracy):", svm_cv.best_score_)
```

```python
#TASK 7: Calculate SVM Accuracy on Test Data
#NHIỆM VỤ 7: Tính toán độ chính xác của SVM trên dữ liệu thử nghiệm
accuracy_svm = svm_cv.score(X_test, Y_test)
print("SVM Test Accuracy:", accuracy_svm)
```

```python
#TASK 8: Decision Tree Classifier Hyperparameter Tuning
#NHIỆM VỤ 8: Điều chỉnh siêu tham số của bộ phân loại cây quyết định
from sklearn.tree import DecisionTreeClassifier

# Create Decision Tree model
tree = DecisionTreeClassifier(random_state=2)

# Define hyperparameter grid
# Xác định lưới siêu tham số
parameters_tree = {
    'max_depth': [2, 4, 6, 8, None],
    'min_samples_split': [2, 5, 10, 20],
    'criterion': ['gini', 'entropy']
}

# Create GridSearchCV object with 10-fold CV
tree_cv = GridSearchCV(tree, parameters_tree, cv=10)

# Fit on training data
tree_cv.fit(X_train, Y_train)

# Display best parameters and accuracy
print("Best Decision Tree parameters:", tree_cv.best_params_)
print("Best Decision Tree score (CV accuracy):", tree_cv.best_score_)
```

```python
#TASK 9: Decision Tree Accuracy on Test Data
#NHIỆM VỤ 9: Độ chính xác của cây quyết định trên dữ liệu thử nghiệm
# Accuracy of Decision Tree on test set
# Độ chính xác của Cây quyết định trên tập kiểm tra
accuracy_tree = tree_cv.score(X_test, Y_test)
print("Decision Tree Test Accuracy:", accuracy_tree)
```

```python
#TASK 10: K-Nearest Neighbors (KNN) Hyperparameter Tuning
from sklearn.neighbors import KNeighborsClassifier

# Create KNN model
knn = KNeighborsClassifier()

# Define hyperparameter grid
parameters_knn = {
    'n_neighbors': [3, 5, 7, 9, 11],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan']
}

# Create GridSearchCV object with 10-fold CV
knn_cv = GridSearchCV(knn, parameters_knn, cv=10)

# Fit on training data
knn_cv.fit(X_train, Y_train)

# Display best parameters and CV score
print("Best KNN parameters:", knn_cv.best_params_)
print("Best KNN CV accuracy:", knn_cv.best_score_)
```

```python
#TASK 11: KNN Accuracy on Test Data
#NHIỆM VỤ 11: Độ chính xác của KNN trên dữ liệu thử nghiệm
# Accuracy of KNN on test set
accuracy_knn = knn_cv.score(X_test, Y_test)
print("KNN Test Accuracy:", accuracy_knn)
```

```python
#TASK 12: Find the Best Model

#We compare all test accuracies to see which model performs best on unseen data:

# Compare accuracies
accuracies = {
    'Logistic Regression': accuracy_logreg,
    'SVM': accuracy_svm,
    'Decision Tree': accuracy_tree,
    'KNN': accuracy_knn
}

best_model = max(accuracies, key=accuracies.get)
print("Best performing model on test data:", best_model)
print("Accuracy:", accuracies[best_model])
```
