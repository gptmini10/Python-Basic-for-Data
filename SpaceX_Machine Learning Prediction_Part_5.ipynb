```python
def plot_confusion_matrix(y, y_predict):
    "This function plots the confusion matrix"
    from sklearn.metrics import confusion_matrix

    cm = confusion_matrix(y, y_predict)
    ax = plt.subplot()
    sns.heatmap(cm, annot=True, ax=ax)  # annot=True để hiển thị số trong ô
    ax.set_xlabel('Predicted labels')
    ax.set_ylabel('True labels')
    ax.set_title('Confusion Matrix')
    ax.xaxis.set_ticklabels(['did not land', 'land'])
    ax.yaxis.set_ticklabels(['did not land', 'landed'])
```

```python
import pandas as pd

URL1 = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv"
URL2 = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_3.csv"
```

```python
data = pd.read_csv(URL1)
X = pd.read_csv(URL2)
```

```python
display(data.head())
display(X.head(100))
```

```python
Task 1: Create a NumPy array from the column Class in data
Applying the method to_numpy() then assign it to the variable Y
```

```python
Nhiệm vụ 1: Tạo một mảng NumPy từ cột Class trong data
Áp dụng phương thức to_numpy() rồi gán nó cho biến Y
```

```python
Y = data['Class'].to_numpy()

print(Y[:10])  
```

```python
Task 2: Standardize the data
```

```python
Nhiệm vụ 2: Chuẩn hóa dữ liệu
```

```python

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X = scaler.fit_transform(X)
print(X.shape)
print(X[:5])
```

```python

TASK3:  create training and testing data
Use the function train_test_split to split the data X and Y into training and test data. Set the parameter test_size to 0.2 and random_state to 2.
The training data and test data should be assigned to the following labels: X_train, X_test, Y_train, Y_test
```

```python
NHIỆM VỤ 3: Tạo dữ liệu huấn luyện và kiểm tra
Sử dụng hàm train_test_split để chia dữ liệu X và Y thành dữ liệu huấn luyện và dữ liệu kiểm tra. Đặt tham số test_size thành 0,2 và random_state thành 2.
Dữ liệu huấn luyện và dữ liệu kiểm tra nên được gán các nhãn sau: X_train, X_test, Y_train, Y_test
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=2
)

print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("Y_train shape:", Y_train.shape)
print("Y_test shape:", Y_test.shape)
```

```python

TASK 4: Find Hyperparameter
Create a Logistic Regression object.
Create a GridSearchCV object (logreg_cv) with cv = 10.
Fit the model to find the best parameters from the dictionary parameters.
```

```python
NHIỆM VỤ 4: Tìm Siêu tham số
Tạo một đối tượng Hồi quy Logistic.
Tạo một đối tượng GridSearchCV (logreg_cv) với cv = 10.
Điều chỉnh mô hình để tìm ra các tham số tốt nhất từ ​​các tham số từ điển.
```

```python
parameters = {
    'C': [0.01, 0.1, 1],
    'penalty': ['l2'],
    'solver': ['lbfgs']
}
```

```python
# Create Logistic Regression object
lr = LogisticRegression(max_iter=1000)

# Create GridSearchCV object with 10-fold cross-validation
logreg_cv = GridSearchCV(lr, parameters, cv=10)

# Fit the model to find the best parameters
logreg_cv.fit(X_train, Y_train)
```

```python
print("Tuned hyperparameters (best parameters):", logreg_cv.best_params_)
print("Best accuracy:", logreg_cv.best_score_)
```
